# 立讯精密收购分析

## 联滔电子

```python
type = 'ProfitStatement'
code = '002075'
url = 'http://money.finance.sina.com.cn/corp/go.php/vDOWN_{}/displaytype/4/stockid/{}/ctrl/all.phtml'.format(type, code)
url
```

```python
html = get_html_text(url)
```

```python
from io import StringIO
origin_df =pd.read_table(StringIO(html))
```

```python
df = origin_df.T
df.columns = df.iloc[0]
df.drop(list(df.index[[0,-2,-1]]), axis=0, inplace=True)
df.drop('单位', axis=1, inplace=True)
```

```python
df = df.astype('float64')
df.index = pd.to_datetime(df.index)
```

```python
plt.figure()
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False 
df[['应收票据', '应收账款']].plot()
```
